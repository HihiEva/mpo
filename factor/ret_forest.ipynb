```python
import pandas as pd
import os

def load_factor_csvs(csv_dir):
    """
    从目录加载所有CSV文件到pandas DataFrame的字典中。

    Args:
        csv_dir (str): 包含CSV文件的目录路径。

    Returns:
        dict: 一个字典，其中键是文件名（不带扩展名），值是相应的DataFrame。
    """
    dfs = {}

    if not os.path.isdir(csv_dir):
        print(f"错误：找不到目录 {csv_dir}")
        return dfs

    print(f"正在从 {csv_dir} 加载文件...")

    for filename in os.listdir(csv_dir):
        if filename.endswith('.csv'):
            df_name = 'df_' + os.path.splitext(filename)[0]
            file_path = os.path.join(csv_dir, filename)
            
            try:
                
                # 读取CSV，将第一列解析为日期时间并设为索引
                # 移除已废弃的 date_parser，改用 date_format='mixed' 自动兼容推断多种日期格式
                df = pd.read_csv(file_path, index_col=0, parse_dates=True, date_format='mixed')
                

                
                # 在全局变量中创建DataFrame
                globals()[df_name] = df
                print(f" - 已成功加载 {filename} 到名为 '{df_name}' 的DataFrame。")

            except Exception as e:
                print(f" - 加载 {filename} 时出错: {e}")
    
    return dfs

if __name__ == '__main__':
    # 用户提供的路径
    factor_dir = r'C:\Users\CtaBackup\Desktop\fac\Factor_CSVs'
    
    # 加载DataFrames
    load_factor_csvs(factor_dir)

    print("脚本执行完毕。如果CSV文件被成功加载，您现在应该可以在Jupyter Notebook或IPython环境中直接通过带有 'df_' 前缀的文件名访问对应的DataFrame。")
    print("例如，如果有名为 '001.csv' 的文件，您可以直接使用变量 `df_001`。")
```

```python
import pandas as pd
import os
import glob

# 设置路径
folder_path = r'C:\Users\CtaBackup\Desktop\py_data\day1_2'

# 方法1: 使用 glob 读取所有 CSV 文件
csv_files = glob.glob(os.path.join(folder_path, "*.csv"))

# 定义标准列名
standard_columns = [
    'date',                # 对应 day1(:, 1)
    'open',                # 对应 day1(:, 2)
    'high',                # 对应 day1(:, 3)
    'low',                 # 对应 day1(:, 4)
    'close',               # 对应 day1(:, 5)
    'volume',              # 对应 day1(:, 6)
    'turnover',            # 对应 day1(:, 7)
    'open_interest',       # 对应 day1(:, 8)
    'volume_total',        # 对应 day1(:, 9)
    'open_interest_total', # 对应 day1(:, 10)
    'carry',               # 对应 day1(:, 11)
    'close_raw',           # 对应 day1(:, 12)
    'carry2',              # 对应 day1(:, 13)
    'receipt',             # 对应 day1(:, 14)
    'limit_up',            # 对应 day1(:, 15)
    'limit_down',          # 对应 day1(:, 16)
    'hycc_long_top5',      # 对应 day1(:, 17)
    'hycc_short_top5',     # 对应 day1(:, 18)
    'hycc_volume_top5',    # 对应 day1(:, 19)
    'spread1'              # 对应 day1(:, 20)
]
# 创建字典存储所有数据
data_dict = {}
for file in csv_files:
    filename = os.path.basename(file).replace('.csv', '')
    
    # 添加错误处理，跳过空文件
    try:
        df = pd.read_csv(file)
        
        # 检查DataFrame是否为空
        if df.empty or len(df.columns) == 0:
            print(f"跳过空文件: {filename}")
            continue
        
        # 根据实际列数调整列名
        actual_cols = min(len(df.columns), len(standard_columns))
        new_columns = standard_columns[:actual_cols] + df.columns[actual_cols:].tolist()
        
        # 重命名列
        df.columns = new_columns
        data_dict[filename] = df
        
        print(f"已加载: {filename}, 形状: {df.shape}, 列名: {list(df.columns)}")
    
    except pd.errors.EmptyDataError:
        print(f"跳过空文件: {filename}")
        continue
    except Exception as e:
        print(f"读取文件 {filename} 时出错: {str(e)}")
        continue

print(f"总共加载了 {len(data_dict)} 个CSV文件")

```

# 降频

```python
import pandas as pd
import numpy as np

# =========================================================
# 1. 自动抓取全局环境中的因子 DataFrame
# =========================================================
df_names = sorted([name for name in globals() if name.startswith('df_') and name[3:].isdigit()])

print(f"✅ 成功找到 {len(df_names)} 个高频因子，开始执行【标准商品期货交易日】降频...")

# =========================================================
# 2. 批量循环处理并在内存中生成新变量
# =========================================================
for name in df_names:
    try:
        # 获取原始的 15 分钟频数据
        df_15m = globals()[name].copy()
        
        # 将字符串时间索引转化为 Pandas 时间戳
        if not pd.api.types.is_datetime64_any_dtype(df_15m.index):
            df_15m.index = pd.to_datetime(df_15m.index)
            
        # 确保时间轴是升序排列的 (保证 .last() 取到的是真正的收盘)
        df_15m = df_15m.sort_index()
            
        # =====================================================
        # 🌟 终极修复：标准交易日划分法则
        # =====================================================
        # 将原有的自然时间统一向后平移 4 小时，然后抹去具体时间只留日期。
        # 此时前一天 21:00 和当天的 15:00，都会得到同一个 "当天的日期" 标签！
        trading_date = (df_15m.index + pd.Timedelta(hours=4)).normalize()
        
        # -----------------------------------------------------
        # 后缀 _1：日内均值法 (Mean) - 融合夜盘与日盘的全天共识
        # -----------------------------------------------------
        # 用 groupby 替代 resample，按我们新算出来的交易日标签分组
        df_daily_mean = df_15m.groupby(trading_date).mean().dropna(how='all')
        globals()[f"{name}_1"] = df_daily_mean  
        
        # -----------------------------------------------------
        # 后缀 _2：尾盘信号法 (Last) - 捕捉下午 15:00 收盘前的终局意图
        # -----------------------------------------------------
        # 因为我们事先 sort_index 过，这里取出的必定是下午 15:00 (或各品种最后的交易时间)
        df_daily_last = df_15m.groupby(trading_date).last().dropna(how='all')
        globals()[f"{name}_2"] = df_daily_last  
        
    except Exception as e:
        print(f"❌ 处理 {name} 时出错: {e}")

print("=========================================================")
print(f"🎉 降频彻底完成！现已完美适配国内商品【T-1 21:00 至 T 15:00】的交易日规则！")
print("👉 方法1 (均值法): df_001_1, df_002_1 ... df_514_1")
print("👉 方法2 (尾盘法): df_001_2, df_002_2 ... df_514_2")
```

# 数据清理
未上市保留nan，直接就是不加到树模型的回归中

```python
import pandas as pd
import numpy as np

print("=========================================================")
print("🚀 第一阶段 (定制版)：全局变量抓取、掩码清洗与截面压平")
print("=========================================================\n")

# =========================================================
# 1. 自动抓取内存中所有的 _1 和 _2 因子变量
# =========================================================
# 抓取逻辑：变量名以 'df_' 开头，且以 '_1' 或 '_2' 结尾
factor_names = sorted([name for name in globals() if name.startswith('df_') and (name.endswith('_1') or name.endswith('_2'))])

print(f"0. 成功从内存中抓取到 {len(factor_names)} 个高频降频因子！")

# 1. 统一基准的时间轴格式为 Datetime
print("\n1. 正在统一时间轴为 Datetime 格式...")
returns.index = pd.to_datetime(returns.index.astype(str)).normalize()
valid_trading_mask.index = pd.to_datetime(valid_trading_mask.index.astype(str)).normalize()

# 收集所有因子的时间索引
all_indices = [returns.index, valid_trading_mask.index]
for name in factor_names:
    df = globals()[name]
    df.index = pd.to_datetime(df.index.astype(str)).normalize()
    all_indices.append(df.index)

# 取交集
common_dates = returns.index
for idx in all_indices:
    common_dates = common_dates.intersection(idx)
common_dates = sorted(common_dates)
print(f"   ✅ 对齐完成，共有 {len(common_dates)} 个有效交易日交集。")

ret_aligned = returns.loc[common_dates]
mask_aligned = valid_trading_mask.loc[common_dates]

# =========================================================
# 2. 🌟 核心：直接套用 Volume 掩码
# =========================================================
print("\n2. 正在应用 Volume 掩码 (将无成交量的假信号彻底踩死)...")
stacked_features = []
actual_feature_names = []

for name in factor_names:
    # 截取共同时间段
    temp_df = globals()[name].loc[common_dates]
    
    # 🌟 掩码大招：有成交量 (True) 的地方保留原信号，没成交量的地方强制转成 NaN
    masked_df = temp_df.where(mask_aligned)
    
    # 3. 顺手把它暴力压平 (Stacking) 
    s = masked_df.stack(dropna=False) 
    s.name = name
    stacked_features.append(s)
    actual_feature_names.append(name)

print("   ✅ 掩码应用与 1D 压平完成！")

# =========================================================
# 4. 组装 X (特征) 和 y (收益)
# =========================================================
print("\n3. 正在组装 X_panel (特征池) 和 y_panel (收益率基准)...")
# 横向拼接所有 94 个特征
X_panel = pd.concat(stacked_features, axis=1)

# 把收益率也压平。注意：我们不动它里面的 NaN，等 7 步预测时用 shift 自动对齐！
y_panel = ret_aligned.stack(dropna=False)
y_panel.name = 'target_return'


# =========================================================
# 5. 彻底去空 (只清洗特征矩阵 X)
# =========================================================
print("\n4. 正在执行特征池清洗 (剔除无成交量的脏数据)...")
# 我们只对 X 执行 dropna，只要某个因子因为掩码变成了 NaN，这天的这个品种就不具备预测资格
master_X_clean = X_panel.dropna() 

clean_samples = len(master_X_clean)
raw_samples = len(X_panel)
drop_ratio = (raw_samples - clean_samples) / raw_samples if raw_samples > 0 else 0

print(f"   ✅ 清洗彻底完成！")
print(f"   - 原始特征池总行数 (天数×品种): {raw_samples}")
print(f"   - 剔除无成交量/无效的行数:      {raw_samples - clean_samples} (占比 {drop_ratio:.2%})")
print(f"   - 🎯 最终纯净可用的特征样本数: {clean_samples} 行")

# 导出供下游使用的干净变量
X_clean_base = master_X_clean
y_clean_base = y_panel
```

```python
import pandas as pd
import numpy as np
import time
from tqdm.auto import tqdm
import pickle
import os

from sklearn.ensemble import RandomForestRegressor
from lightgbm import LGBMRegressor
from xgboost import XGBRegressor
from catboost import CatBoostRegressor

print("=========================================================")
print("🚀 第三阶段：全量滚动 7 步预测与 Alpha 矩阵生成")
print("=========================================================\n")

# =========================================================
# 1. 核心参数设置 (可按需调整)
# =========================================================
lookback = 252                    # 滚动训练窗口天数
horizons = [1, 2, 3, 4, 5, 6, 7]  # 预测的未来跨度 (tau)

# 💡 算力保护提示：为了全量回测能在一个下午跑完，这里 n_estimators 设为 50。
# 如果你想追求极致性能且不在乎等待时间，可以全部改为 100。
# 另外，RF_MAE 极其耗时，如果跑得太慢，建议先将其注释掉。
final_models = {
    'LGBM_MSE': LGBMRegressor(n_estimators=50, max_depth=3, objective='regression', n_jobs=-1, random_state=42, verbose=-1),
    'LGBM_Huber': LGBMRegressor(n_estimators=50, max_depth=3, objective='huber', alpha=1.2, n_jobs=-1, random_state=42, verbose=-1),
    'XGB_MSE': XGBRegressor(n_estimators=50, max_depth=3, objective='reg:squarederror', tree_method='hist', n_jobs=-1, random_state=42),
    'XGB_Huber': XGBRegressor(n_estimators=50, max_depth=3, objective='reg:pseudohubererror', tree_method='hist', n_jobs=-1, random_state=42),
    'Cat_RMSE': CatBoostRegressor(iterations=50, depth=3, loss_function='RMSE', random_seed=42, verbose=0),
    'RF_MSE': RandomForestRegressor(n_estimators=50, criterion='squared_error', max_depth=3, n_jobs=-1, random_state=42),
    # 'RF_MAE': RandomForestRegressor(n_estimators=50, criterion='absolute_error', max_depth=3, n_jobs=-1, random_state=42), # 极慢，慎开
}

# =========================================================
# 2. 预先构建所有 tau 的目标收益率表 (提速优化)
# =========================================================
print("1. 正在为 7 个预测跨度构建时空平移目标...")
y_wide = y_clean_base.unstack()
y_targets_1d = {}

for tau in horizons:
    # 把未来的收益率拉到今天
    shifted_wide = y_wide.shift(-tau)
    y_targets_1d[tau] = shifted_wide.stack(dropna=False)
    
all_dates = sorted(X_clean_base.index.get_level_values(0).unique())
total_days = len(all_dates)

# 用于存储最终预测结果的大字典
# 格式: all_predictions[model_name][(date, tau)] = pd.Series(...)
all_predictions = {model_name: {} for model_name in final_models.keys()}

print(f"   ✅ 构建完毕！总共有 {total_days} 个交易日参与滚动。")


# =========================================================
# 3. 开启时间轴滚动循环 (Rolling)
# =========================================================
print(f"\n2. 🚀 正式启动滚动训练，窗口={lookback}天，请耐心等待...")

# 从 lookback + max(horizons) 天开始，确保第一次训练有足够的数据
start_idx = lookback + max(horizons)

for i in tqdm(range(start_idx, total_days), desc="时间滚动进度"):
    current_date = all_dates[i]
    
    # 提取今天的特征 (用于预测)
    # X_today 是一个 DataFrame，索引是品种名，列是 94 个因子
    X_today = X_clean_base.loc[current_date] 
    
    # 遍历所有要预测的期限 tau
    for tau in horizons:
        
        # 🌟 核心防偷窥逻辑 (防未来函数)：
        # 要预测 tau 天后，最新的【已兑现目标】只能是 current_date - tau 那天的。
        # 因此，训练集的最晚一天必须是 current_date - tau。
        train_end_idx = i - tau
        train_start_idx = train_end_idx - lookback
        
        train_dates = all_dates[train_start_idx : train_end_idx]
        
        # 切片获取训练集的 X
        X_train_chunk = X_clean_base.loc[train_dates]
        
        # 切片获取训练集的 y (对应 tau 跨度的特定目标)
        y_train_chunk = y_targets_1d[tau].loc[train_dates]
        
        # 严格对齐 X 和 y，剔除 y 为 NaN 的行
        train_df = pd.concat([X_train_chunk, y_train_chunk], axis=1, join='inner').dropna()
        
        if len(train_df) < 100:
            # 如果有效样本太少，跳过这天的这个 tau 预测
            continue
            
        X_train_ready = train_df[X_train_chunk.columns].astype(float)
        y_train_ready = train_df.iloc[:, -1].astype(float)
        
        # 遍历所有模型进行训练和预测
        for model_name, model in final_models.items():
            try:
                # 训练模型
                model.fit(X_train_ready, y_train_ready)
                
                # 预测未来的收益
                preds = model.predict(X_today.astype(float))
                
                # 组装为带品种索引的 Series
                pred_series = pd.Series(preds, index=X_today.index, dtype=float)
                
                # 存入 MPO 字典格式 [(t, tau)]
                all_predictions[model_name][(current_date, tau)] = pred_series
                
            except Exception as e:
                pass # 忽略偶发的拟合失败

print("\n🎉 所有的 7 步滚动预测已经全部完成！")

```

```python
import pandas as pd
import numpy as np
import quant_eval
from IPython.display import display

print("=========================================================")
print("🏆 第四阶段：模型评测与图表输出 (Evaluation & Output)")
print("=========================================================\n")

print(">>> 1. 正在将预测结果重构为 quant_eval 兼容的 T x N (列表) 矩阵格式...")
models_to_compare = {}
max_h = max(horizons) # max(horizons) 通常是 7

for model_name, mpo_dict in all_predictions.items():
    if not mpo_dict: # 如果某个模型没跑出数据，跳过
        continue
        
    # 临时字典用于组装列表
    date_asset_dict = {}
    
    for (date, tau), series in mpo_dict.items():
        if date not in date_asset_dict:
            date_asset_dict[date] = {}
            
        for asset, val in series.items():
            if pd.isna(val): 
                continue # 忽略空值
            if asset not in date_asset_dict[date]:
                # 初始化一个长度为 7 的全 NaN 列表
                date_asset_dict[date][asset] = [np.nan] * max_h
            # 把 tau 的预测值塞进列表对应的位置 (tau=1 放 index 0)
            date_asset_dict[date][asset][tau - 1] = val
            
    # 将字典转换为 Date x Asset 的 DataFrame，单元格里全是 [pred_1, ..., pred_7]
    df_list_format = pd.DataFrame.from_dict(date_asset_dict, orient='index')
    df_list_format.sort_index(inplace=True)
    models_to_compare[model_name] = df_list_format

print(f"   ✅ 重构完成！当前参与回测的树模型数量: {len(models_to_compare)} 个\n")

# =========================================================
# 2. 调用回测引擎
# =========================================================
target_h = 1  # 设定你想评估的跨度，如果要看 T+5，这里改成 5

print(f">>> 2. 开始进行多空组合回测与画图 (评估 T+{target_h} 期)...")

# 注意：为了绝对的时间轴对齐，我们传入第一阶段清洗好的 ret_aligned 
backtest_stats = quant_eval.evaluate_portfolio_backtest(
    models_dict=models_to_compare,
    true_df=ret_aligned,        
    h=target_h,                 # quant_eval 内部会去取列表里的第 h 项
    top_pct_extreme=0.1,        # 极值组比例
    top_pct_moderate=0.3        # 温和组比例
)

# =========================================================
# 3. 显示最终的核心回测指标对比表
# =========================================================
print(f"\n=== 📊 树模型核心指标回测结果对比 (预测 T+{target_h} 期) ===")
# 按照 Sharpe Ratio 降序排列，谁是冠军一目了然！
sorted_stats = backtest_stats.sort_values(by='Sharpe', ascending=False)
display(sorted_stats.style.format("{:.4f}"))
```

```python
import quant_eval
import importlib
from IPython.display import display

# 强制重新加载以应用刚保存的新函数
importlib.reload(quant_eval)

print("=========================================================")
print("🔬 深度体检：模型在不同时间切片下的稳定性对比")
print("=========================================================\n")

# =========================================================
# 1. 核心参数设置
# =========================================================
target_h = 1      # 展现预测第几步的结果 (h=1 为预测次日)
n_splits = 5      # 把整个时间段切成几份 (比如 5 份代表 5 个子周期)

print(f">>> 正在将全样本切分为 {n_splits} 个阶段，评估 T+{target_h} 期预测...")

# 调用我们刚写好的切分评估函数
split_report = quant_eval.evaluate_split_periods(
    models_dict=models_to_compare,  # 这里的字典就是我们在第四阶段生成的那个大全集
    true_df=ret_aligned, 
    h=target_h, 
    n_splits=n_splits
)

# =========================================================
# 2. 定制极性高亮渲染器 (横向色阶对比)
# =========================================================
def style_split_report(report_df):
    """
    为多层报告自动填充色系：
    - 颜色越深，代表这个模型在这一行的表现越优！
    """
    # 找出所有的指标名称
    metrics = report_df.index.get_level_values('Metric').unique()
    
    styler = report_df.style.format("{:.4f}")
    
    for metric in metrics:
        # 选出属于该指标的所有行作为应用颜色的子集 (Mask)
        idx_mask = report_df.index.get_level_values('Metric') == metric
        
        # 判断是不是“越小越好”的误差类指标
        if metric in ['MAE', 'RMSE']:
            # 使用 Reds_r 反转红色系：数值越小（越优秀），颜色越红！
            styler = styler.background_gradient(axis=1, subset=(idx_mask, report_df.columns), cmap='Reds_r')
        else:
            # 其他“越大越好”的胜率/IC类指标
            # 使用 Blues 蓝色系：数值越大（越优秀），颜色越蓝！
            styler = styler.background_gradient(axis=1, subset=(idx_mask, report_df.columns), cmap='Blues')
            
    return styler

# =========================================================
# 3. 渲染出图
# =========================================================
print("\n=== 分段稳定性评估表 (一排之内：颜色最深的列 = 性能冠军) ===")
styled_table = style_split_report(split_report)
display(styled_table)
```
